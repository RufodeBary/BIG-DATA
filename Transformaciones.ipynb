##Transformación, join de datasets

```python

# Notas de Matemática y Lengua
df1 = pd.DataFrame({'Alumno': ['Alumno0', 'Alumno1', 'Alumno2'],
                    'Matermática': ['5', '6', '7'],
                    'Lengua': ['7', '7', '8']
                    })

# Notas de Historia y Geografía
df2 = pd.DataFrame({'Alumno': ['Alumno0', 'Alumno2', 'Alumno3'],
                    'Historia': ['4', '5', '4'],
                    'Geografía': ['6', '7', '9']
                    })

# Mostramos los dataframes
print(df1)
print('           ')
print(df2)

```

```python
# Hacemos join de los dataframes y mostramos las notas

merged_df = pd.merge(df1, df2, on='Alumno', how='left')
print(merged_df)
```

##Eliminar columnas que no son necesarias

```python

# Creamos un DataFrame de ejemplo
data = {'nombre': ['Ana', 'Pedro', 'María'],
        'edad': [30, 25, 32],
        'ciudad': ['Madrid', 'Barcelona', 'Sevilla'],
        'correo': ['ana@gmail.com', 'pedro@gmail.com', 'maria@gmail.com'],
        'telefono': [123456789, 987654321, 567891234]}
df = pd.DataFrame(data)

print(df)
```

```python
# Eliminar las columnas 'correo' y 'telefono'
columnas_a_eliminar = ['correo', 'telefono']

#axis=1: Se refiere a las columnas.
#inplace=True: modifica el DataFrame original en lugar de crear una copia y realizar los cambios en ella
df.drop(columnas_a_eliminar, axis=1, inplace=True)

print(df)
```

##Agrupamiento

```python
# Creamos un DataFrame de ventas
data = {'Vendedor': ['Ana', 'Pedro', 'Ana', 'María', 'Pedro', 'Ana'],
        'Monto': [100, 200, 150, 300, 100, 50],
        'Operaciones': [1, 1, 2, 1, 3, 1]}
df_ventas = pd.DataFrame(data)

print("Datos originales:")
print(df_ventas)
```

```python
# Agrupamos por 'Vendedor' y sumamos sus montos
# El proceso es: 1. Agrupar, 2. Seleccionar columna, 3. Aplicar función (sum, mean, count)
ventas_totales = df_ventas.groupby('Vendedor')['Monto'].sum().reset_index()

print("\nVentas totales por vendedor:")
print(ventas_totales)
```

## Conversión de Mayúsculas

```python
# Creamos un DataFrame con nombres variados
data = {'Estudiante': ['ana gomez', 'PEDRO RODRIGUEZ', 'mAría loPez']}
df_nombres = pd.DataFrame(data)

print("Datos originales:")
print(df_nombres)
```

```python
# Convertir a MAYÚSCULAS
df_nombres['Estudiante_Mayus'] = df_nombres['Estudiante'].str.upper()
print(df_nombres)
```

```python
# Convertir a minúsculas
df_nombres['Estudiante_Minus'] = df_nombres['Estudiante'].str.lower()
print(df_nombres)
```

1. Transformación de Fechas y Horas: Un cargamento de datos llega siempre exactamente 7 días después de la fecha de hoy. Crea un script que obtenga la fecha actual y le sume una semana (7 días), mostrando el resultado en formato día/mes/año

Pista: Usa la librería datetime y timedelta

```python
#Ejercicio 1
import datetime
# Obtener la fecha actual
fecha_actual = datetime.date.today()

# Sumar 7 días a la fecha actual
dias_a_sumar = datetime.timedelta(days=7)
fecha_futura = fecha_actual + dias_a_sumar

# Formatear la fecha futura a 'día/mes/año'
fecha_formateada = fecha_futura.strftime('%d/%m/%Y')

# Imprimir el resultado
print(f"La fecha actual es: {fecha_actual.strftime('%d/%m/%Y')}")
print(f"El cargamento llegará el: {fecha_formateada}")
```

3. Filtrado de Datos: Dado el siguiente DataFrame, filtra los datos para mostrar solo las ventas que superan los 500 y son del producto 'A'.

```python
#Ejercicio 3
data = {'producto': ['A', 'B', 'A', 'C'], 'ventas': [600, 450, 750, 300]}
df = pd.DataFrame(data)
print("DataFrame original:")
print(df)
print('\n')

# Filtrar los datos: ventas > 500 y producto == 'A'
df_filtrado = df[(df['ventas'] > 500) & (df['producto'] == 'A')]

print("Ventas filtradas (producto 'A' y ventas > 500):")
print(df_filtrado)


```

5. Unión de DataFrames: Tienes dos DataFrames, uno con datos de empleados y otro con sus salarios. Únelos para que en un solo DataFrame aparezca el nombre del empleado y su salario.
